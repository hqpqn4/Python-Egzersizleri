```python
# VKI hesaplama
print '*' * 15, "VKI Hesaplama Aracı", '*' * 15
ad = raw_input("Adiniz: ")
boy = raw_input("Boyunuzu (m) girin: ")
agirlik = raw_input("Kilonuzu (kg) girin: ")

print "Sayin %s girmis oldugunuz degerlere gore; " %ad
vki = float(agirlik) / (float(boy) * float(boy))

print "Vucut kitle endeksiniz: %.2f" %vki

if vki < 16:
    print "VKI Araliginiz: Ileri duzeyde zayif"
elif 16 < vki < 16.99:
    print "VKI Araliginiz: Orta duzeyde zayif"
elif 17 < vki < 18.49:
    print "VKI Araliginiz: Hafif duzeyde zayif"
elif 18.50 < vki < 24.99:
    print "VKI Araliginiz: Normal kilolu"
elif 25 < vki < 29.99:
    print "VKI Araliginiz: Hafif sisman"
elif 30 < vki < 34.99:
    print "VKI Araliginiz: 1. dereceden obez"
elif 35 < vki < 39.99:
    print "VKI Araliginiz: 2. dereceden obez"
else:
    print "VKI Araliginiz: 3. dereceden obez"
```
