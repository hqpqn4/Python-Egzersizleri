```python
import os
```

```python
# binary dosya kopyalama
def kopyala(dosya1, dosya2):
    d1 = open(dosya1, "r+b")
    d2 = open(dosya2, "r+b")
    
    okunan = d1.read()
    d2.write(okunan)
    
    d1.close()
    d2.close()
```

```python
kopyala("carpim_tablosu.jpg", "resim2.jpg")
```
